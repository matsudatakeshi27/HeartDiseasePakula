```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('RawData/Cleveland Imputed Data.csv')
print(df)
train, test = train_test_split(df, test_size=0.33)
print(train)
print(test)
train.to_csv("NewTrainTestData/cle_train.csv",index = False)
test.to_csv("NewTrainTestData/cle_test.csv",index = False)
```
