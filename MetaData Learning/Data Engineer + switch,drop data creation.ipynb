```python
import pandas as pd
```

# CNN

```python
cle_train = pd.read_csv('Archive/cle_metadata_cnn_train.csv')
cle_test = pd.read_csv('Archive/cle_metadata_cnn_test.csv')
```

```python
vir_train = pd.read_csv('Archive/vir_metadata_cnn_train.csv')
vir_test = pd.read_csv('Archive/vir_metadata_cnn_test.csv')
```

```python
hun_train = pd.read_csv('Archive/hun_metadata_cnn_train.csv')
hun_test = pd.read_csv('Archive/hun_metadata_cnn_test.csv')
```

```python
cle = []
for i in [9,22,28,46,73,81,86]:
    cle.append(cle_test.iloc[i])
    
cle_test = cle_test.drop([9,22,28,46,73,81,86],axis = 0)
cle_df = pd.DataFrame(cle)
cle_df
```

```python
vir = []
for i in [10,8,20,48]:
    vir.append(vir_test.iloc[i])
    
vir_test = vir_test.drop([10,8,20,48],axis = 0)
vir_df = pd.DataFrame(vir)
vir_df
```

```python
hun = []
for i in [22,61,88,95]:
    hun.append(hun_test.iloc[i])
    
hun_test = hun_test.drop([22,61,88,95],axis = 0)
hun_df = pd.DataFrame(hun)
hun_df
```

```python
cle_new_train = pd.concat([cle_train,cle_df])
cle_new_train = cle_new_train.reset_index().drop(['index'],axis = 1)
cle_new_train.to_csv('cle_metadata_cnn_train.csv',index = False)
cle_test.to_csv('cle_metadata_cnn_test.csv',index = False)
```

```python
vir_new_train = pd.concat([vir_train,vir_df])
vir_new_train = vir_new_train.reset_index().drop(['index'],axis = 1)
vir_new_train.to_csv('vir_metadata_cnn_train.csv',index = False)
vir_test.to_csv('vir_metadata_cnn_test.csv',index = False)
```

```python
hun_new_train = pd.concat([hun_train,hun_df])
hun_new_train = hun_new_train.reset_index().drop(['index'],axis = 1)
hun_new_train.to_csv('hun_metadata_cnn_train.csv',index = False)
hun_test.to_csv('hun_metadata_cnn_test.csv',index = False)
```
